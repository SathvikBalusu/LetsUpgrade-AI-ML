**Range()**
range() function is used to give values in that specified range.
It returns a sequence of numbers starting from 0 by default and increments by 1 until specified number. 


```
#demonstrating a program using range()
for i in range(5):
  print(i)
print("values from starting index to ending index-1 is printed")
for i in range(1,6):
  print(i)
print("values from starting index to ending index-1 with stepsize is printed")
#stepsize value skips to that number in the sequence
for i in range(1,10,2):
  print(i)
```
