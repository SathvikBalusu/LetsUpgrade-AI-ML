```
#subtraction of two complex numbers
a= 2+3j
b= 1+2j
print(a-b)
```

```
#4th root of a number 
a=81 
b=a**(1/4)
print("the 4th root of",a,"is",b)
```

```
#Swapping two numbers using temporary variable
a=2
b=3
print("before swapping",a,b )
temp=a
a=b
b=temp
print("after swapping",a,b)
```

```
#swapping two numbers without temp variable
a=2
b=3
print("before swapping",a,b)
a=a+b
b=a-b
a=a-b
print("after swapping",a,b  )
```

```
#program to convert fahrenheit to kelvin and celsius
f=100 
print("fahrenheit is",f)
k=273.15+(f-32)*5/9
print(" kelvin of ",f,"fahrenheit is",k)
c=(f-32)*5/9
print(" celsius of",f,"fahrenheit is",c)
```
