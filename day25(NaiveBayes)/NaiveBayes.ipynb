```python
import numpy as np
import pandas as pd
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import *
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
data=pd.read_csv(r"C:\Users\sathv\LU\datasets\train.csv")
```

```python
data.head()
```

```python
data1=data.drop(['Cabin','Ticket','Name','PassengerId'],axis=1)
```

```python
data1
```

```python
le=preprocessing.LabelEncoder()
data1['Sex']=le.fit_transform(data1['Sex'])
data1['Embarked']=le.fit_transform(data1['Embarked'])
```

```python
IDV=['Survived', 'Pclass', 'Sex',  'SibSp', 'Parch','Embarked']
```

```python
def NaiveBayes(name):
    y=data1[name]
    X=data1.drop(name,axis=1)
    X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0)
    clf=BernoulliNB()
    y_pred=clf.fit(X_train,y_train).predict(X_test)
    print(f"Accuracy Score and Confusion matrix of {name} with other Independent variables")
    print(accuracy_score(y_test,y_pred,normalize=True))
    cm=confusion_matrix(y_test,y_pred)
    print(cm)
    print("The total correct values predicted are ",np.trace(cm),"out of total values ",np.sum(cm),"giving accuracy of ",np.trace(cm)/np.sum(cm))
    print("\n\n")
```

```python
for name in IDV:
    NaiveBayes(name)
```
