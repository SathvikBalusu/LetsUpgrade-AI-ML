```
#Palindrome or anagram
str1="racecar"
str2="carrace"
def palindrome(str):
   if str==str[::-1]:
     print("the string",str,"is palindrome")
   else: 
     print("the string",str, "is not a palindrome")
def anagram(str1,str2):
  l1=len(str1)
  l2=len(str2)
  if(l1!=l2):
    return 0
  str1=sorted(str1)
  str2=sorted(str2)
  for i in range(0,l1):
    if(str1[i]!=str2[i]):
       return 0
  else:
    return 1
palindrome(str1)
palindrome(str2)
if(anagram(str1,str2)):
  print("the two strings are anagram")
else:
  print("the two strings are not anagram")
```
