```python
import numpy as np
import pandas as pd
from sklearn import preprocessing
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn import svm
from sklearn.model_selection import train_test_split
```

```python
data=pd.read_csv(r"C:\Users\sathv\LU\datasets\train.csv")
data.head()
```

```python
le=preprocessing.LabelEncoder()
data['Sex']=le.fit_transform(data['Sex'])
data['Embarked']=le.fit_transform(data['Embarked'])
data=data.drop(['Ticket','Cabin','PassengerId','Name'],axis=1)
```

```python
data.head()
```

```python
DV=['Survived','Pclass','Sex','SibSp','Parch','Embarked']
```

```python
def SVM(name):
    y=data[name]
    x=data.drop(name,axis=1)
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
    clf=svm.SVC(gamma=0.01,C=100)
    clf.fit(x_train,y_train)
    y_pred=clf.predict(x_test)
    print(f"Accuracy score and Confusion matrix with {name} and other Independent Variables is:")
    print(accuracy_score(y_test,y_pred,normalize=True))
    cm=confusion_matrix(y_test,y_pred)
    print(cm)
    print("The correct values predicted are",np.trace(cm),"out of total values",np.sum(cm),"Giving accuracy of",np.trace(cm)/np.sum(cm),"\n\n")
```

```python
for name in DV:
    SVM(name)
```
